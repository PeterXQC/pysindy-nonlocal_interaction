```python
import glob
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Get a list of all files matching the pattern *.npy
files = glob.glob('*.npy')

# Load each file into a list using numpy.load()
data = [np.load(file) for file in files]
```

We first plot two of the spatial dimensions at a specific time t and feature in a rasterized plot. We expect there to be a block where there is value, and everywhere else should be zero.

```python
# plot the matrix as a heatmap
print(np.shape(data[0]))
plt.imshow(data[0][:, :, 37, 14], cmap='hot', interpolation='nearest')
plt.show()
```
